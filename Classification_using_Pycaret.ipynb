```python
# check installed version
import pycaret
pycaret.__version__
```

## Load Data

```python
# loading sample dataset from pycaret dataset module
from pycaret.datasets import get_data
data = get_data('income')
```

## Glass having pottatium- unique Entries

```python
print(data['education'].unique())
```

## Setup Classification

```python
# import pycaret classification and init setup
from pycaret.classification import *
s = setup(data, target = 'education', session_id = 123)
```

## Import Classification Experiment

```python
# import ClassificationExperiment and init the class
from pycaret.classification import ClassificationExperiment
exp = ClassificationExperiment()
```

## Setup

```python
exp.setup(data, target = 'education', session_id = 123)
```

## Compare Models

```python
# compare baseline models
best = compare_models()
```

## Compare models

```python
# compare models using OOP
 exp.compare_models()
```

## Plot

```python
# plot confusion matrix
plot_model(best, plot = 'confusion_matrix')
```

## Plot AUC

```python
# plot AUC
plot_model(best, plot = 'auc')
```

## plot feature importance

```python
# plot feature importance
plot_model(best, plot = 'feature')
```

```python
# check docstring to see available plots
help(plot_model)
```

## Evaluate Model


```python
evaluate_model(best)
```

## Predict

```python
# predict on test set
holdout_pred = predict_model(best)
```

## Show Prediction

```python
# show predictions df
holdout_pred.head()
```

## New Data set

```python
# copy data and drop Class variable

new_data = data.copy()
new_data.drop('education', axis=1, inplace=True)
new_data.head()
```

## Prediction on new Data set

```python
# predict model on new_data
predictions = predict_model(best, data = new_data)
predictions.head()
```

## Save Pipline

```python
# save pipeline
save_model(best, 'prediction_model')
```
